Univariate analysis

Boxplot untuk mendeteksi outlier

```python
# correlation heatmap
correlation = healthy.corr()
sns.heatmap(correlation, annot=True, fmt='.2f')
```

**Note **

From correlation map highest positive correlation between life expectancy (years) and Happiness Levels so these features have high corelation. more higher happiness level, more higher life expectancy.  others factors like outdoor activities have little correlation with another features. 

```python
df = healthy.groupby(['City']).agg(
    pollution_score=('Pollution(Index score) (City)','sum')
).reset_index().sort_values('pollution_score',ascending=False)
df
```

Note

From this dataframe we can see three top highhest Pollution(Index score)(City) are: Cairo, Beijing, and Jakarta

```python
healthy.groupby(['City']).agg(
    happy_level=('Happiness levels(Country)','median')
).reset_index().sort_values('happy_level',ascending=False)
```

Note

From this dataframe we can see three top highhest Pollution(Index score)(City) are: Cairo, Beijing, and Jakarta

```python
healthy.groupby(['City']).agg(
    life_expect=('Life expectancy(years) (Country)','median')
).reset_index().sort_values('life_expect',ascending=False)
```

Note

From this dataframe we can see three top highest life expect are: Fukuoka,Tokyo, and Milan with age +/- 83 years old. 

```python
healthy.groupby(['City']).agg(
    avg_hours=('Annual avg. hours worked','sum')
).reset_index().sort_values('avg_hours',ascending=False)
```

Note

From this dataframe we can see three top average hours worked are: Mexico,Soul, and Moscow.
